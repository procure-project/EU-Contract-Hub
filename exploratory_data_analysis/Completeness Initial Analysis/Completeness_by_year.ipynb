```python
import os
import pandas as pd


folder_path = 'C:\\Users\\afont\\OneDrive\\Escritorio\\PROCURE\\Data\\Completeness_Data\\OpenTender'


# List all XLSX files in the folder
xlsx_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx')]
# Dictionary to store DataFrames
otender = {}

countries=[]
# Iterate through each XLSX file
for file_name in xlsx_files:
    # Create the full file path
    file_path = os.path.join(folder_path, file_name)
    # Read the XLSX file into a Pandas DataFrame
    df = pd.read_excel(file_path)

    # Use the file name (without extension) as the DataFrame name
    df_name = os.path.splitext(file_name)[0].split('-')[1].upper()
    countries.append(df_name)
    # Add the DataFrame to the dictionary
    otender[df_name] = df.drop(df.index[-1])
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
# Loop through countries and create subplots
for c in countries:
    # Extract the relevant DataFrame for the country
    df_country = otender[c]
    fig, ax = plt.subplots()



    # Line plot for tender_finalPrice_currency
    ax.plot(df_country['year'], df_country['tender_cpvs'], marker='o', linestyle='-', color='lightsteelblue', label='CPVs')
    # Line plot for tender_mainCpv
    plt.plot(df_country['year'], df_country['tender_mainCpv'], marker='o', linestyle='-', color='b', label='Main CPV')

    # Line plot for tender_finalPrice_currency
    ax.plot(df_country['year'], df_country['bid_price_EUR'], marker='o', linestyle='-', color='lightcoral', label='Bid Price')
    # Line plot for tender_finalPrice_currency
    ax.plot(df_country['year'], df_country['tender_finalPrice_currency'], marker='o', linestyle='-', color='r', label='Final Price')



    #Threshold
    ax.axhline(y=100, color='black', linestyle=':')

    # Set axis limits
    ax.set_ylim(0, 100)

    # Add labels and title to each subplot
    ax.set_ylabel('Complete entries (%)')
    ax.set_xlabel('Year')
    ax.set_title(f'{c} - OpenTender Completeness')
    ax.legend(loc='upper left', bbox_to_anchor=(0, 1), prop={'size': 8})
    ax.set_ylim(0, 125)
    imgname='C:\\Users\\afont\\OneDrive\\Escritorio\\PROCURE\\Data\\Completeness_Data\\Images\\'+c+'_opentender.png'
    fig.savefig(imgname, dpi=300)  # Specify the filename and optionally the dpi (dots per inch)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import os
import pandas as pd
folder_path = 'C:\\Users\\afont\\OneDrive\\Escritorio\\PROCURE\\Data\\Completeness_Data\\TED'


# List all XLSX files in the folder
xlsx_files = [file for file in os.listdir(folder_path) if file.endswith('.xlsx')]

# Dictionary to store DataFrames
ted = {}

countries=[]
# Iterate through each XLSX file
for file_name in xlsx_files:
    # Create the full file path
    file_path = os.path.join(folder_path, file_name)
    # Read the XLSX file into a Pandas DataFrame
    df = pd.read_excel(file_path)

    # Use the file name (without extension) as the DataFrame name
    df_name = os.path.splitext(file_name)[0].split('-')[1]
    countries.append(df_name)
    # Add the DataFrame to the dictionary
    ted[df_name] = df.drop(df.index[-1])
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
# Loop through countries and create subplots
for c in countries:
    # Extract the relevant DataFrame for the country
    df_country = ted[c]
    fig, ax = plt.subplots()

    # Line plot for tender_mainCpv
    plt.plot(df_country['year'], df_country['CPV'], marker='o', linestyle='-', color='b', label='CPV')

    # Line plot for tender_finalPrice_currency
    ax.plot(df_country['year'], df_country['AWARD_VALUE_EURO'], marker='o', linestyle='-', color='r', label='Final Price')

    #Threshold
    ax.axhline(y=90, color='black', linestyle=':')

    # Set axis limits
    ax.set_ylim(0, 100)

    # Add labels and title to each subplot
    ax.set_ylabel('Complete entries (%)')
    ax.set_xlabel('Year')
    ax.set_title(f'{c} - TED Completeness')
    ax.legend()
    imgname='C:\\Users\\afont\\OneDrive\\Escritorio\\PROCURE\\Data\\Completeness_Data\\Images\\'+c+'_TED.png'
    fig.savefig(imgname, dpi=300)  # Specify the filename and optionally the dpi (dots per inch)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
